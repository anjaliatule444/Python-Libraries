```python
## Imports
import matplotlib.pyplot as plt
```

### Pie Chart

```python
## Data creation
areas=['Marketing','Sales','Developement','HR','Customer Support']
budget=[2.5,4,10,1,6]
```

```python
plt.pie(x=budget,labels=areas)
```

```python
## With plt.show() function all the data that was appearing before pie chart will get removed and only graph will be visible.
plt.pie(x=budget,labels=areas,autopct='%0.2f%%')
plt.show()
```

0.2 will show the percentage upto 2 decimal points.

### Scatter Plot

```python
x=[23,45,12,49,97,32,11,44]
y=[45,34,67,87,55,22,89,90]
plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary')
```

### 2D Line Plot

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[233,456,770,120,222,444,200,300]
offline_price=[400,300,100,333,444,565,565,899]
chor_bazaar=[100,200,300,400,150,250,450,500]

plt.plot(product_id,online_price,color='orange',marker='D',markersize=5,label='Online Prices')
plt.plot(product_id,offline_price,color='red',markersize=5,label='Offline Prices')
plt.plot(product_id,chor_bazaar,color='green',marker='+',markersize=5,label='Chor bazar Prices')
plt.xlabel("Product")
plt.ylabel("Price")
plt.title("Product vs Price")
plt.legend() ## To display labels--> It will find the empty space on its own and display the labels
plt.show()
```

Marker=D stands for diamond. With above graph we can decide which product from where we can buy i.e. offline,online or from chor bazaar.

### Bar Graphs

```python
year=[2012,2013,2014,2015,2016,2017,2018]
price=[65000,69000,75000,55000,52000,64000,81000]

plt.bar(year,price,color='black')
plt.xlabel("Year")
plt.ylabel("Price In INR")
plt.title("Year vs Price")
plt.show()
```

```python
year=[2012,2013,2014,2015,2016,2017,2018]
price=[65000,69000,75000,55000,52000,64000,81000]

plt.barh(year,price,color='black')
plt.xlabel("Price In INR")
plt.ylabel("Year")
plt.title("Year vs Price")
plt.show()
```
